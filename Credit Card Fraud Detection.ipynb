```python
data=data.drop('Time',axis=1)
```

```python
duplicate_rows = data.duplicated()
duplicate_data = data[duplicate_rows]
print(duplicate_data)
```

```python
data=data.drop_duplicates()
```

```python
#Standardization of the Amount Column
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
data['Amount']=sc.fit_transform(data[['Amount']])
```

```python
#Visulization of count of Class
sb.countplot(data['Class'])
```

```python
#Taking dependent and independent variable
x=data.drop('Class',axis=1)
y=data['Class']
```

```python
#oversampling for imbalance dataset
from imblearn.over_sampling import SMOTE
```

```python
sm=SMOTE()
```

```python
x_res,y_res=sm.fit_resample(x,y)
```

```python
y_res.value_counts()
```

```python
#Dividing the dataset into train and test
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x_res,y_res,test_size=0.3,random_state=42)
```

```python
#Performing the dataset into ml model
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
```

```python
rfc.fit(xtrain,ytrain)
```

```python
rfc.score(xtest,ytest)
```

```python
ypred=rfc.predict(xtest)
```

```python
#Checking accuracy,f1,precission,recall
from sklearn.metrics import classification_report,accuracy_score
```

```python
report=classification_report(ytest,ypred)
```

```python
print(report)
```

```python
accuracy_score(ytest,ypred)
```
